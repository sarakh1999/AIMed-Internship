plot 2d funcion: x^4 + 3x^3 - x^2 + 5

```python
x = np.linspace(-1,1, 1000) # from -1 to 1 and 1000 samples total
f = np.power(x,4) + np.power(x,3) *3 + np.power(x,2) * (-1) + 5
plt.plot(x,f)
```

mutiple plots using subplots
plot four different functions and adjust spacing between them
add title for each subplots
add title for the plot
function1: f(X) = x

function2: f(X) = x^2

function3: f(X) = (x-1)^2

function4: f(X) = x^3

```python
x = np.linspace(-1,1, 1000) # from -1 to 1 and 1000 samples total
fig, ax =  plt.subplots(2,2,figsize=(10,10))

ax[0][0].plot(x,x)
ax[0][0].set_title('x')

ax[0][1].plot(x,np.power(x,2))
ax[0][1].set_title('x^2')

ax[1][0].plot(x,np.power((x-1),2))
ax[1][0].set_title('(x-1)^2')

ax[1][1].plot(x,np.power(x,3))
ax[1][1].set_title('x^3')
```
