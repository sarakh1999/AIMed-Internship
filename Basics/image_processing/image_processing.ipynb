### Manual histogram for each color channel

```python
def show_histogram(img):
    img = img.flatten()
    histogram = np.full(256, 0)
    for intensity in img:
        histogram[intensity]+=1
    pdf = histogram / sum(histogram)
    plt.ylabel('Frequency')
    plt.xlabel('Intensity');
    return pdf

r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
red = plt.hist(x=np.arange(256), weights= show_histogram(r), bins=256, density = True, color = "red")
green = plt.hist(x=np.arange(256), weights= show_histogram(g), bins=256, density = True, color = "green")
blue = plt.hist(x=np.arange(256), weights= show_histogram(b), bins=256, density = True, color = "blue")
```

### Built-in python histogram function

```python
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

rotate image 60 degrees

```python
height, width = img.shape[:2]
# get the center of the image
center_x, center_y = (width/2, height/2)
# rotate the image by 60 degrees counter-clockwise around the center of the image
M = cv2.getRotationMatrix2D((center_x, center_y), 60, 1.0)
rotated_image = cv2.warpAffine(img, M, (width, height))
plt.imshow(rotated_image)
```

resize images to 600x600 and show both images

```python
resized = cv2.resize(img, (600, 600) , interpolation = cv2.INTER_AREA)
plt.imshow(resized);
```

```python
resized = cv2.resize(rotated_image, (600, 600) , interpolation = cv2.INTER_AREA)
plt.imshow(resized);
```
